```python
import networkx as nx
```

```python
G = nx.Graph()   # 生成空图
Node = nx.path_graph(15)  # 生成15个节点的随机无向图
G.add_nodes_from(Node)  #将这15个节点加入图中
```

```python
import random
```

```python
def rand_edge(vi,vj,p=0.8):		#默认概率p=0.1
    probability =random.random()#生成随机小数
    if(probability>p):			#如果大于p
        print(probability)
        G.add_edge(vi,vj)       #添加边
```

```python
i = 0
while i<15:
    j = 0
    while j<15:
        if j<i:
            j+=1
        else:
            rand_edge(i,j)
            #print(i,j)
            j+=1
    i+=1
```

```python
import matplotlib.pyplot as plt
pos = nx.circular_layout(G)
options = {
    "with_labels": True,
    "font_size": 10,
    "font_weight": "bold",
    "font_color": "white",
    "node_size": 500,
    "width": 2
}
nx.draw_networkx(G, pos, **options)
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

```python
# 查看上述随机图5->9的最短路径及长度
print([p for p in nx.all_shortest_paths(G=G,source=5,target=9,method="dijkstra")])
print(nx.shortest_path_length(G=G,source=5,target=9,method="dijkstra"))
```

```python
# 获得5->9之间的所有路径
print([p for p in nx.all_simple_paths(G=G,source=5,target=9)])
```
