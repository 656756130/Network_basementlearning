```python
import networkx as nx
import random
```

```python
G = nx.Graph()   # 生成空图
Node = nx.path_graph(15)  # 生成15个节点的随机无向图
G.add_nodes_from(Node)  #将这15个节点加入图中
```

```python
# 建立一个随机网络
def rand_edge(vi,vj,p=0.8):		#默认概率p=0.1
    probability =random.random()#生成随机小数
    if(probability>p):			#如果大于p
        print(probability)
        G.add_edge(vi,vj)       #添加边
```

```python
i = 0
while i<15:
    j = 0
    while j<15:
        if j<i:
            j+=1
        else:
            rand_edge(i,j)
            #print(i,j)
            j+=1
    i+=1
```

```python
import matplotlib.pyplot as plt
pos = nx.circular_layout(G)
options = {
    "with_labels": True,
    "font_size": 10,
    "font_weight": "bold",
    "font_color": "white",
    "node_size": 500,
    "width": 2
}
nx.draw_networkx(G, pos, **options)
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

```python
#计算13号的集聚系数
print(nx.clustering(G=G,nodes=13))
```

```python
#计算平均集聚系数
print(nx.average_clustering(G=G))
```

```python
#计算全局集聚系数
print(nx.transitivity(G=G))
```
