```python
# 度和度分布
# 度的计算
import networkx as nx
G = nx.Graph()
G.add_nodes_from([1,2,3,4])
G.add_edges_from([(1,2),(1,3),(2,3),(2,4)])
```

```python
nx.draw(G,node_size = 800,with_labels=True)
```

```python
d = nx.degree(G)
print(d)
```

```python
# 将元组变为字典形式方便查询
d = dict(nx.degree(G))
# 整体查询度的值
print(d.values())
# 查询节点名称
print(d.items())
# 查询特定节点的度
print(d[1])
```

```python
import numpy as np
degree_list = np.array(d.values())
print(degree_list)
```

```python
# 获取度分布
Distribution = nx.degree_histogram(G)
print(Distribution)   # 为0-dmax的频数
```

```python
Distribution = np.array(Distribution)
p_k = Distribution/len(G.nodes)
print(p_k)
k = np.array(range(0,len(Distribution)))
print(k)
```

```python
# 画图
import matplotlib.pyplot as plt
plt.bar(k, p_k, width=0.5, color="red")
plt.xlabel("$k$")
plt.ylabel("$p_k$")
plt.xlim([0,4])
```
