```python
# 创建图
import networkx as nx
```

```python
# 创建无向图
# 创建空图
G = nx.Graph()
# 创建1 2 3 4节点
G.add_nodes_from([1,2,3,4])
# 创建链接
G.add_edges_from([(1,2),(1,3),(2,3),(2,4)])
```

```python
nx.draw(G = G,node_size = 600, with_labels= True)
```

```python
# 创建有向图
# 创建空有向图
G_Di = nx.DiGraph()
G_Di.add_nodes_from([1,2,3,4])
# 关系是from to的关系
G_Di.add_edges_from([(1,2),(1,3),(2,3),(2,4)])
```

```python
nx.draw(G = G_Di,node_size = 600, with_labels= True)
```

```python
# 创建加权无向图图
# 创建空图
G_Wei = nx.Graph()
G_Wei.add_weighted_edges_from([(1,2,2.5),(1,3,2.0),(2,3,1.5),(2,4,1.1)])
# 展示邻接矩阵
matrix = nx.adjacency_matrix(G_Wei)
print(matrix)
nx.draw(G = G_Wei,node_size = 600,with_labels=True)
```

```python
# 将邻接矩阵变为致密矩阵形式
dense_matrix = matrix.todense()
print(dense_matrix)
```

```python
# 通过邻接矩阵来创建图
import numpy as np
known_matrix = np.array([[0,1,1,0],[1,0,1,1],[1,1,0,0],[0,1,0,0]])
G_fromMatrix = nx.from_numpy_matrix(known_matrix)
nx.draw(G = G_fromMatrix,node_size = 600,with_labels=True)
```
