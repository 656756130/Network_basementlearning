```python
import random_networks as rn
```

```python
G = rn.empty_network(10)
print(G)
```

```python
rn.create_random_networks(G=G,nodes_number=10)
```

```python
import matplotlib.pyplot as plt
pos = rn.nx.circular_layout(G)
options = {
    "with_labels": True,
    "font_size": 10,
    "font_weight": "bold",
    "font_color": "white",
    "node_size": 500,
    "width": 2
}
rn.nx.draw_networkx(G, pos, **options)
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```
